```python
import pandas as pd
```

```python
df_dados = pd.read_csv("base_leite_2023_2024.csv", encoding="utf-8")
df_dados
```

```python
colunas_numericas = list(df_dados.columns)[5:]
colunas_numericas
```

```python
for column in colunas_numericas:
    df_dados[column] = df_dados[column].astype(str).str.replace(r'\.(?=[^.]*$)', ',', regex=True)
    df_dados[column] = df_dados[column].str.replace('.', '').replace(",", ".")
    df_dados[column] = df_dados[column].str.replace(',', '.').astype(float)

df_dados
```

```python
df_dados.to_excel("Base Leite 2023-2024 Formatada.xlsx", index=False)
```
